# COVID-19 and Pneumonia Deaths in the State of Arizona (2020-2023)

```python
# importing the required libraries and uploading the dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataframe = pd.read_csv(r'c:\Users\Jordan\Downloads\Data Projects\Datasets\C19_Deaths_AZ.csv')

dataframe.head(5)
```

```python
# pulling up the basic outline of the data

dataframe.info()
```

```python
# pulling some basic statistics

dataframe.describe()
```

```python
# visualizing the relationship between age and C19 deaths

plt.bar (x=dataframe['Age_Group'], height=dataframe['COVID19_Deaths'])

plt.plot()

plt.xlabel('Age Group')
plt.ylabel('Deaths')
plt.title('COVID-19 Deaths by Age')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

## Observations on Age Groups (COVID-19)

There have been approximately 30,000 deaths in the state of Arizona as the result of COVID-19.

The age group with the most deaths is the 75 - 84 years old category.

```python
# visualizing the relationship between sex and C19 deaths

plt.bar (x=dataframe['Sex'], height=dataframe['COVID19_Deaths'])

plt.plot()

plt.xlabel('Sex')
plt.ylabel('Deaths')
plt.title('COVID-19 Deaths by Sex')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

## Observations on Sex (COVID-19)

Men are more likely to die from COVID-19 than women.  

```python
# visualizing the relationship between age and pneumonia deaths

plt.bar (x=dataframe['Age_Group'], height=dataframe['Pneumonia_Deaths'])

plt.plot()

plt.xlabel('Age Group')
plt.ylabel('Deaths')
plt.title('Pneumonia Deaths by Age')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

## Observations on Age Group (Pneumonia)

The most vulnerable population is still the 75 - 84 years old category.

```python
# visualizing the relationship between sex and pneumonia deaths

plt.bar (x=dataframe['Sex'], height=dataframe['Pneumonia_Deaths'])

plt.plot()

plt.xlabel('Sex')
plt.ylabel('Deaths')
plt.title('Pneumonia Deaths by Sex')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

## Observations on Sex (Pneumonia)

Men are more likely to die from pneumonia than women.
